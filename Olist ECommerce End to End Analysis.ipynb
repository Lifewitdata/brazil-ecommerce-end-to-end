```python
reference_date = df['order_purchase_timestamp'].max() + timedelta(days=1)

rfm = df.groupby('customer_unique_id').agg(
    Recency   = ('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    Frequency = ('order_id', 'nunique'),
    Monetary  = ('price', 'sum')
).reset_index()
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'],   q=4, labels=[4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first'),  q=4, labels=[1, 2, 3, 4])

rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

```python
def assign_segment(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])
    if r == 4 and f == 4:
        return 'Champion'
    elif r >= 3 and f >= 3:
        return 'Loyal Customer'
    elif r == 4:
        return 'Recent Customer'
    elif r == 2 and f >= 2:
        return 'At Risk'
    elif r <= 1:
        return 'Lost'
    else:
        return 'Potential Loyalist'
rfm['Segment'] = rfm.apply(assign_segment, axis=1)      
```

```python
seg_counts = rfm['Segment'].value_counts().reset_index()
seg_counts.columns = ['Segment', 'Count']

fig_rfm = px.bar(seg_counts, 
                 x='Segment', y='Count',
                 color='Segment',
                 title='Customer Segmentation (RFM Analysis)',
                 text='Count',
                 color_discrete_sequence=px.colors.qualitative.Bold)

fig_rfm.update_traces(textposition='outside')
fig_rfm.update_layout(showlegend=False)
fig_rfm.show()
```

```python
# ── Scikit-learn ──────────────────────────────────────────────
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix,
    accuracy_score, roc_auc_score, roc_curve
)

# ── XGBoost ───────────────────────────────────────────────────
from xgboost import XGBClassifier

print(' All libraries imported successfully!')
```

```python
# ── 1.2  Merge into one dataframe ─────────────────────────────
df = (orders
      .merge(items,     on='order_id',   how='left')
      .merge(customers, on='customer_id',how='left')
      .merge(products,  on='product_id', how='left')
      .merge(payments,  on='order_id',   how='left')
      .merge(reviews,   on='order_id',   how='left'))

# ── 1.3  Parse dates ──────────────────────────────────────────
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for c in date_cols:
    df[c] = pd.to_datetime(df[c])
```

```python
# ── 1.4  Basic cleaning ───────────────────────────────────────
df['review_comment_title']   = df['review_comment_title'].fillna('No Title')
df['review_comment_message'] = df['review_comment_message'].fillna('No Message')
df['product_category_name']  = df['product_category_name'].fillna('other')
df.dropna(subset=['product_id','price','payment_type'], inplace=True)

for c in ['product_weight_g','product_length_cm','product_height_cm','product_width_cm']:
    df[c] = df[c].fillna(df[c].median())

for c in ['product_name_lenght','product_description_lenght','product_photos_qty']:
    df[c] = df[c].fillna(0)
```

```python
# ── 1.5  Derived time features ────────────────────────────────
df['delivery_time_days'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days
df['delivery_diff_days'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days
df['purchase_month'] = df['order_purchase_timestamp'].dt.month
df['purchase_hour']  = df['order_purchase_timestamp'].dt.hour
df['year_month']     = df['order_purchase_timestamp'].dt.to_period('M')

print(f' dataframe ready  →  {df.shape[0]:,} rows  ×  {df.shape[1]} columns')
```

```python
# delivery_diff_days > 0  →  delivered AFTER estimated  →  LATE  (1)
# delivery_diff_days <= 0 →  delivered ON or BEFORE     →  ON TIME (0)

df1 = df.dropna(subset=['delivery_diff_days']).copy()
df1['is_late'] = (df1['delivery_diff_days'] > 0).astype(int)

print('Target distribution:')
counts = df1['is_late'].value_counts()
print(f'  On Time (0): {counts[0]:,}  ({counts[0]/len(df1)*100:.1f}%)')
print(f'  Late    (1): {counts[1]:,}  ({counts[1]/len(df1)*100:.1f}%)')
```

```python
# Features we will use to predict delay
features_m1 = [
    'product_weight_g',       # heavier → maybe slower?
    'freight_value',          # shipping cost proxy
    'price',                  # item value
    'product_length_cm',      # size
    'product_height_cm',
    'product_width_cm',
    'purchase_month',         # seasonality
    'purchase_hour',
    'customer_state',         # geography
    'product_category_name'   # category behaviour
]

X1 = df1[features_m1].copy()
y1 = df1['is_late']

# Encode categorical columns with LabelEncoder
le = LabelEncoder()
X1['customer_state']        = le.fit_transform(X1['customer_state'].astype(str))
X1['product_category_name'] = le.fit_transform(X1['product_category_name'].astype(str))

print('Feature matrix shape:', X1.shape)
X1.head(3)
```

```python
# 80% training, 20% testing  |  random_state for reproducibility
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42, stratify=y1
)

print(f'Training samples : {len(X1_train):,}')
print(f'Testing  samples : {len(X1_test):,}')
```

```python
# ── Logistic Regression (needs scaling) ───────────────────────
scaler = StandardScaler()
X1_train_sc = scaler.fit_transform(X1_train)
X1_test_sc  = scaler.transform(X1_test)

lr1 = LogisticRegression(max_iter=500, random_state=42)
lr1.fit(X1_train_sc, y1_train)
lr1_pred = lr1.predict(X1_test_sc)
lr1_acc  = accuracy_score(y1_test, lr1_pred)

# ── Random Forest ─────────────────────────────────────────────
rf1 = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf1.fit(X1_train, y1_train)
rf1_pred = rf1.predict(X1_test)
rf1_acc  = accuracy_score(y1_test, rf1_pred)

# ── XGBoost ───────────────────────────────────────────────────
xgb1 = XGBClassifier(n_estimators=100, random_state=42,
                     eval_metric='logloss', verbosity=0)
xgb1.fit(X1_train, y1_train)
xgb1_pred = xgb1.predict(X1_test)
xgb1_acc  = accuracy_score(y1_test, xgb1_pred)

print('=== MODEL 1 — Delivery Delay — Accuracy Comparison ===')
print(f'  Logistic Regression : {lr1_acc*100:.2f}%')
print(f'  Random Forest       : {rf1_acc*100:.2f}%')
print(f'  XGBoost             : {xgb1_acc*100:.2f}%  ← usually best')
```

```python
# Full classification report for best model (XGBoost)
print('=== XGBoost — Classification Report ===')
print(classification_report(y1_test, xgb1_pred, target_names=['On Time','Late']))

# Confusion matrix heatmap
cm1 = confusion_matrix(y1_test, xgb1_pred)
fig_cm1 = px.imshow(
    cm1,
    labels=dict(x='Predicted', y='Actual', color='Count'),
    x=['On Time','Late'], y=['On Time','Late'],
    text_auto=True, color_continuous_scale='Blues',
    title='Confusion Matrix — Delivery Delay (XGBoost)'
)
fig_cm1.show()
```

```python
FORECAST_MONTHS = 6

# Create future dates (monthly frequency)
future = model_prophet.make_future_dataframe(periods=FORECAST_MONTHS, freq='MS')
forecast = model_prophet.predict(future)

# Show only the forecast period
print('=== Forecast for next 6 months ===')
forecast_tail = forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(FORECAST_MONTHS)
forecast_tail.columns = ['Month','Forecast','Lower Bound','Upper Bound']
forecast_tail['Forecast']    = forecast_tail['Forecast'].round(2)
forecast_tail['Lower Bound'] = forecast_tail['Lower Bound'].round(2)
forecast_tail['Upper Bound'] = forecast_tail['Upper Bound'].round(2)
print(forecast_tail.to_string(index=False))
```
